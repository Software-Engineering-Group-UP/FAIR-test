```python
import pandas as pd

# Load the CSV file
file_path = 'combined.csv'  # replace with the path to your file
data = pd.read_csv(file_path)

# Print column count and column names
column_count = len(data.columns)
print(f"Number of Columns: {column_count}")
print(f"Column Names: {data.columns.tolist()}")

# Display the first 5 rows
data.head()
```

```python
from collections import Counter

# Combine all rows into a single string
all_text = ' '.join(data['language'].dropna())

# Split the text into words and count occurrences
word_count = Counter(all_text.split())

# Display the word count
print(word_count)
```
